#### **Moving window:**
1. rolling window: focus more on recent obs, i.e., how last week data tell?
2. expanding window: long last obs share equal weights as latest obs, and all obs (life time) at hand are used to observe pattern or trend

```python
ts = pd.Series(np.random.randn(20), pd.date_range('4/7/20', freq = 'D', periods = 20))
ts
```

```python
ts_lagged = ts.shift()
ts_lagged
```

```python
plt.figure(figsize = (15, 6))
plt.plot(ts, color = 'blue')
plt.plot(ts_lagged, color = 'red')
plt.show()
```

```python
ts_h = pd.Series(np.random.randn(20), pd.date_range('4/7/20', freq = 'H', periods = 20))
ts_h_lagged = ts_h.shift()
ts_h_lagged
```

```python
plt.figure(figsize = (15, 6))
plt.plot(ts_h, color = 'blue')
plt.plot(ts_h_lagged, color = 'red')
plt.show()
```

#### **A stock data and practice:**

```python
import os
from datetime import datetime
import pandas_datareader.data as web

start = pd.Timestamp('2010-1-1')
end = pd.Timestamp('2013-1-27')
f = web.DataReader("F", 'yahoo', start, end)
```

```python
f.head(3)
```

```python
f_0708_2012 = f['2012-07' : '2012-08']
```

```python
f_0708_2012[['High', 'Low']].plot()
plt.show()
```

```python
r = f.rolling(50).var()['Volume'].plot()
plt.show()
```

```python
r = f.expanding().var()['Volume'].plot()
plt.show()
```

##### **On how many days did the stock close higher than it opened?**

```python
len(f[f['Close'] > f['High']])             # wrong, got no idea how
```

```python
lagged = f.shift(1)
sum((f - lagged)['Open'] > 0)
```

##### **And did the 'up' days become more or less frequent over time?**

```python
f['DayGain'] = f['Open'] - lagged['Open']
sum(f['DayGain'] > 0)/len(f['DayGain'])
```

```python
f.rolling(window = 25)['DayGain'].apply(lambda x: len([x_i for x_i in x if x_i > 0])/len(x)).plot()
plt.show()
```

```python
f.rolling(window = 100)['DayGain'].apply(lambda x: len([x_i for x_i in x if x_i > 0])/len(x)).plot()
plt.show()
```

```python
f.expanding()['DayGain'].apply(lambda x: len([x_i for x_i in x if x_i > 0])/len(x)).plot()
plt.show()
```

```python
f['High'].resample('M').mean().rolling(window = 1).mean().plot()
plt.show()
```

```python
f.resample('M').mean()['High'].plot()
plt.show()
```

```python
volume = f.Volume
volume_lagged = f.Volume.shift()
diffed_volume = volume - volume_lagged
diffed_volume.rolling(window = 20).var().plot()
plt.show()
```

```python
diffed_volume_2 = f.Volume.diff()
diffed_volume_2.rolling(window = 20).var().plot()
plt.show()
```
